```python
df.drop([7,48,54,113,63],inplace = True)   # dropping those countries whose complete data were not found.
df.reset_index()                           # these countries included New Caledonia, Kosovo, Curacao, Gibraltar, Northern Mariana islands etc 


# some of the data were wrongly filled and were corrected
# df.loc[df['alcohol consumption in Liter'] == '$34,299']                             
# df['alcohol consumption in Liter'].replace({"$34,299": "8.4 litres"}, inplace=True)



# df.loc[df['alcohol consumption in Liter'] == '$61,063']
# df['alcohol consumption in Liter'].replace({"$61,063": "2.87 litres"}, inplace=True)


```

```python
df['price_of_a_pint_of_beer'] = df.price_of_a_pint_of_beer.apply(lambda x: x.replace('$',''))                   # replacing the string to change it to float

df['alcohol consumption in Liter'] = df['alcohol consumption in Liter'].apply(lambda x: x.replace('litres','')) # replacing the string to change it to float
df['alcohol consumption in Liter'] = df['alcohol consumption in Liter'].apply(lambda x: x.replace('<',''))
df['alcohol consumption in Liter'] = df['alcohol consumption in Liter'].astype(float) # changing the datatype to perform the calculation

df['price_of_a_pint_of_beer'] = df['price_of_a_pint_of_beer'].astype(float)
df['price_of_a_liter_of_beer($)'] = df['price_of_a_pint_of_beer'] / 0.568261 # change pint of beer to liter of beer
df
```

```python
# adding new feature

df['money_spent_on_alcohol($)'] = df['alcohol consumption in Liter'] * df['price_of_a_liter_of_beer($)']
df
```

```python
df['GDP per capita\xa0 in 2020'] = df['GDP per capita\xa0 in 2020'].apply(lambda x: x.replace(',',''))
df['GDP per capita\xa0 in 2020'] = df['GDP per capita\xa0 in 2020'].astype(float)
```

```python
# another feature

df['% of GDP per capita spent on alcohol'] = (df['money_spent_on_alcohol($)'] / df['GDP per capita\xa0 in 2020']) * 100
df
```

```python
# percentile ranking

df['Percentile Rank in %'] = (df['% of GDP per capita spent on alcohol'].rank(pct = True)) * 100
df
```

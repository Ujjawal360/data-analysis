```python

df['price_of_a_pint_of_beer'] = df.price_of_a_pint_of_beer.apply(lambda x: x.replace('$',''))
df['alcohol consumption in Liter'] = df['alcohol consumption in Liter'].apply(lambda x: x.replace('litres',''))
df['alcohol consumption in Liter'] = df['alcohol consumption in Liter'].apply(lambda x: x.replace('<',''))
# df['alcohol consumption in Liter'] = df['alcohol consumption in Liter'].apply(lambda x: x.replace(',',''))
# df['alcohol consumption in Liter'] = df['alcohol consumption in Liter'].apply(lambda x: x.replace('$',''))
df['alcohol consumption in Liter'] = df['alcohol consumption in Liter'].astype(float)
df['price_of_a_pint_of_beer'] = df['price_of_a_pint_of_beer'].astype(float)
df['price_of_a_liter_of_beer($)'] = df['price_of_a_pint_of_beer'] / 0.568261
df
```

```python
df['money_spent_on_alcohol($)'] = df['alcohol consumption in Liter'] * df['price_of_a_liter_of_beer($)']
df
```

```python
df['GDP per capita\xa0 in 2020'] = df['GDP per capita\xa0 in 2020'].apply(lambda x: x.replace(',',''))
df['GDP per capita\xa0 in 2020'] = df['GDP per capita\xa0 in 2020'].astype(float)
```

```python
df['% of GDP per capita spent on alcohol'] = (df['money_spent_on_alcohol($)'] / df['GDP per capita\xa0 in 2020']) * 100
df
```

```python
df['Percentile Rank in %'] = (df['% of GDP per capita spent on alcohol'].rank(pct = True)) * 100
df
```
